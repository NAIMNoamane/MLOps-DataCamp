## 2. Computing duration
### What's the standard deviation of the trips duration in January?

```python
jan_taxi['duration'] = abs(jan_taxi['tpep_pickup_datetime']-jan_taxi['tpep_dropoff_datetime'])
jan_taxi['duration_minute'] = jan_taxi['duration'].dt.total_seconds()/60
jan_taxi['duration_minute']
```

```python
jan_taxi['duration_minute'].std()
```

The standard deviation of trips duration in january is 42.59

## 3. Dropping outliers
#### Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive)

```python
jan_taxi = jan_taxi[(jan_taxi['duration_minute']>=1) & (jan_taxi['duration_minute']<=60)]
jan_taxi
```
